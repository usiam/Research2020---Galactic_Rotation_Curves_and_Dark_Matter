### Linear function to fit

```python
def f(x, m, b):
    return m*x + b
```

```python
def linear_bestfit(data, galEnvironment):
           
    xdata = np.log10(data['HI' + '-' + galEnvironment])
    ydata = np.log10(data['NSA_Mstar' + '-' + galEnvironment])
    
    popt, pcov = curve_fit(f, xdata, ydata)  
    m_bf, b_bf = popt
    m_err, b_err = np.sqrt(np.diag(pcov))
    
    m_bf_2 = round(m_bf, 2)
    b_bf_2 = round(b_bf, 2)
   
    ydata_line = m_bf * xdata + b_bf
    
    print('Best-fit line equation for ' + galEnvironment + ' galaxies given by the equation: log(Mstar) = ' + str(m_bf_2) + ' log(MHI) + ' + str(b_bf_2))
    
    return ydata_line
```
